# Limpieza del set de datos
En este proceso de limpieza haremos uso de diferentes herramientas de Pandas que nos permitirán abordar las situaciones que más comúnmente encontraremos en un set de datos tabular, como son:

1. El manejo de datos faltantes (que como lo acabamos de ver están presentes en nuestro set de datos)
2. La eliminación de columnas irrelevantes (es decir, columnas que no contienen información relacionada con el problema que queremos resolver)
3. El manejo de registros (filas) repetidos
4. El manejo de valores extremos (u «outliers») para el caso de las variables (columnas) numéricas
5. La limpieza de errores tipográficos que puedan existir en las variables categóricas


# MANEJO DE DATOS FALTANTES

En este caso, en los datos faltantes usaremos un metodo de la libreria sklearn para que nos permitira rellenar los valores faltantes por la media de la columna

```python
import sklearn as sk
imputer = sk.impute.SimpleImputer(strategy='most_frequent')
df[['Gov_Tax']] = imputer.fit_transform(df[['Gov_Tax']])
df[['Education_Campaign']] = imputer.fit_transform(df[['Gov_Tax']])
df.info()
```

```python
variables_numericas = [
  'Year', 'Population', 'Per_Capita_Sugar_Consumption',
  'Total_Sugar_Consumption', 'Sugar_From_Sugarcane', 'Sugar_From_Beet',
  'Sugar_From_HFCS', 'Sugar_From_Other','Avg_Daily_Sugar_Intake',
  'Obesity_Rate','Sugar_Imports', 'Sugar_Exports',
  'Avg_Retail_Price_Per_Kg', 'Gov_Tax','Gov_Subsidies', 'Education_Campaign',
  'Climate_Conditions', 'Sugarcane_Production_Yield'
  ]

variables_categoricas = ['Country', 'Continent']
```

# Eliminacion de valores duplicados

```python
print(f'El set de datos tiene {df.shape[0]} filas y {df.shape[1]} columnas')
df.drop_duplicates(inplace=True)
print(f'El nuevo set de datos tiene {df.shape[0]} filas y {df.shape[1]} columnas')
```

# Manejo de valores atipicos

```python
for variable in variables_numericas:

    fig, ax = plt.subplots()
    plt.boxplot(df[variable], boxprops=dict(color='blue'), whiskerprops=dict(color='red'), capprops=dict(color='green'), medianprops=dict(color='yellow'), flierprops=dict(markerfacecolor='red', marker='o'))
    plt.figure(figsize=(10, 6))
    ax.set_xlabel(variable)
    ax.set_title(variable)
    ax.set_title(f'{variable}')
    plt.show()


```

```python
outliers_tsc = df['Total_Sugar_Consumption']

x_min = outliers_tsc.min()
x_max = outliers_tsc.max()

q1 = outliers_tsc.quantile(0.25)
q2 = outliers_tsc.quantile(0.50)
q3 = outliers_tsc.quantile(0.75)

print(f'Min: {x_min}')
print(f'Q1: {q1}')
print(f'Q2: {q2}')
print(f'Q3: {q3}')
print(f'Max: {x_max}')

print('-'*70)

iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

print(f'Rango intercuartil: {iqr}')
print(f'Limite inferior: {limite_inferior}')
print(f'Limite superior: {limite_superior}')

print('-'*70)

outliers_tsc = outliers_tsc[(outliers_tsc < limite_inferior) | (outliers_tsc > limite_superior)]
print(f'Valore atipicos: {outliers_tsc}')
```

```python
plt.boxplot(df['Total_Sugar_Consumption'], boxprops=dict(color='blue'), whiskerprops=dict(color='red'), capprops=dict(color='green'), medianprops=dict(color='orange'), flierprops=dict(markerfacecolor='red', marker='o'))
plt.show()
```

```python
print(f'dataset antes de eliminar valores atipicos {df.shape[0]} filas y {df.shape[1]} columnas')
df = df[(df['Total_Sugar_Consumption'] >= limite_inferior) & (df['Total_Sugar_Consumption'] <= limite_superior)]
print('-'*70)
print(f'dataset despues de eliminar valores atipicos {df.shape[0]} filas y {df.shape[1]} columnas')
```

```python
plt.boxplot(df['Total_Sugar_Consumption'], boxprops=dict(color='blue'), whiskerprops=dict(color='red'), capprops=dict(color='green'), medianprops=dict(color='orange'), flierprops=dict(markerfacecolor='red', marker='o'))
plt.show()
```

# **¿Por qué América experimentó una disminución en el consumo de azúcar en el año 2002?**

En este apartado, exploraremos las posibles razones por las cuales el consumo de azúcar en América sufrio una disminución en el año 2002 usando los datos que tenemos en el dataset.

```python
df_america_2002 = df_america[(df_america['Year'] == 2002)]
```

```python
gov = ['Gov_Tax', 'Gov_Subsidies', 'Education_Campaign']

for v in gov:
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    valores = df_america_2002[v].value_counts().sort_index()
    etiquetas = ['No', 'Sí']
    plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
    plt.legend()
    plt.title(f'Proporción de {v}')

    plt.subplot(1, 2, 2)
    sns.boxplot(data=df_america_2002, x=v, y='Total_Sugar_Consumption', palette=colores_paises_america, hue='Country')
    plt.title(f'{v}')
    plt.xticks([0, 1], ['No', 'Sí'])

    plt.tight_layout()
    plt.show()
```

A partir de los gráficos, se puede observar como los impuestos, subsidios y campañas educativas del gobierno podrian influir o no en la media del consumo total de azúcar en el año 2002. De este análisis se tienen las siguientes conclusiones:

**1. Impuestos del gobierno sobre el azucar**

  * La mayoría de los países, o sea 81.1% no aplican impuestos sobre el consumo de azúcar, mientras que solo el 18.9% sí aplica impuestos.
  
  * En general, Brasil muestra un consumo alto tanto con impuestos como sin impuestos.

  * En USA y México, la mediana del consumo es similar, pero con impuestos la dispersión disminuye, sugiriendo cierta regulación.

Se puede concluir que aplicar impuestos podría estar asociado con mayor regulación y menos variabilidad en el consumo, tambien cabe acalarar que Brasil se mantiene alto, lo que sugiere que los impuestos podrían no estar generando un efecto muy fuerte allí.

**2. Subsidios Gubernamentales**

  * Solo el 21.6% de los paises cuenta con subsidios, mientras que el 78.4% no.

  * Sin subsidios, Brasil destaca con un consumo muy alto y disperso.

  * Con subsidios, México y USA muestran una mediana baja, mientras que Brasil disminuye su consumo de manera significante.

Se puede concluir que Los subsidios parecen reducir el consumo de azúcar en  países como México y USA, pero en Brasil se observa una caída significativa.
Esto podría indicar que en países con subsidios, el consumo de azúcar está más regulado o menos incentivado.



**3. Campañas Educativas**

  * La distribución es igual que en los impuestos, 81.1% sin campañas educativas y 18.9% con campañas educativas.

  * Brasil muestra altos niveles de consumo con o sin campañas.

  * México y USA muestran medianas similares pero menos dispersión con campañas educativas.
Se puede concluir que las campañas educativas no reducen el consumo total, pero sí podrían estar ayudando a mantenerlo estable.

# **¿Cómo afectan las políticas gubernamentales (impuestos, subsidios, campañas educativas) al consumo de azúcar per cápita?**

```python
def crear_graficos_continentes (df, variable, color):
    plt.figure(figsize=(16, 5))
    plt.subplot(1,2,1)
    sns.boxplot(data=df, x=variable, y='Per_Capita_Sugar_Consumption' , palette=['red', 'green'], hue=variable)
    plt.title(f'{variable}')
    plt.xticks([0, 1], ['No', 'Si'], rotation=90)

    plt.subplot(1, 2, 2)
    counts = df[variable].value_counts()
    labels = ['No', 'Sí']
    plt.pie(
        counts,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        colors=['red', 'green']
    )
    plt.title(f'Proporción de {variable}')
    plt.show()

    plt.figure(figsize=(16, 5))
    plt.subplot(1, 2, 1)
    sns.boxplot(data=df, x=variable, y='Per_Capita_Sugar_Consumption' , palette=color, hue='Country')
    plt.title(f'{variable}')
    plt.xticks([0, 1], ['No', 'Si'], rotation=90)

    plt.subplot(1, 2, 2)
    ax = sns.barplot(data=df, x=variable, y='Per_Capita_Sugar_Consumption', palette=color, hue='Country')

    for i in ax.containers:
      ax.bar_label(i, fmt='%.0f', label_type='edge', padding=3)
    plt.title(f'{variable}')
    plt.xticks([0, 1], ['No', 'Si'], rotation=90)


```

```python
gov = ['Gov_Tax', 'Gov_Subsidies', 'Education_Campaign']

for v in gov:
  crear_graficos_continentes(df_america, v, colores_paises_america)


```

# **¿Qué relación existe entre las condiciones climáticas y el consumo de azúcar per cápita en los diferentes continentes?**

```python
def crear_graficos_climaticos(df, agrupar, colores):
    plt.figure(figsize=(16, 6))

    plt.subplot(1, 2, 1)
    sns.boxplot(data=df, x='Climate_Conditions', y='Per_Capita_Sugar_Consumption', hue=agrupar, palette=colores)
    plt.xlabel('Condiciones climáticas')
    plt.xticks([0, 1, 2, 3, 4], ['1.Tropical', '2.Arido', '3.Templado', '4.Frío', '5.Montaña'])
    plt.ylabel('Consumo de azúcar per cápita (kg)')
    plt.title('Distribución por condiciones climáticas')
    plt.legend(title=agrupar, bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.subplot(1, 2, 2)
    ax = sns.barplot(data=df, x='Climate_Conditions', y='Per_Capita_Sugar_Consumption', hue=agrupar, palette=colores)
    plt.xlabel('Condiciones climáticas')
    plt.xticks([0, 1, 2, 3, 4], ['1.Tropical', '2.Arido', '3.Templado', '4.Frío', '5.Montaña'])
    plt.ylabel('Consumo promedio de azúcar (kg)')
    plt.title('Promedio por condiciones climáticas')
    plt.legend(title=agrupar, bbox_to_anchor=(1.05, 1), loc='upper left')

    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', label_type='edge', padding=3)

    plt.tight_layout()
    plt.show()
```

```python
continentes = df[(df['Continent'] == 'America') | (df['Continent'] == 'Asia') | (df['Continent'] == 'Europe')][['Continent', 'Country', 'Year', 'Per_Capita_Sugar_Consumption', 'Climate_Conditions']].dropna()
colores_continentes = {
    'America': 'red',
    'Asia': 'orange',
    'Europe': 'green'
}

crear_graficos_climaticos(continentes, 'Continent', colores_continentes)

```

* El consumo per cápita de azúcar muestra medianas similares entre Europa, América y Asia a lo largo de las distintas condiciones climáticas. No hay diferencias extremas.

* América y Asia presentan una mayor dispersión en comparación con Europa, especialmente en climas tropicales y áridos.

* No hay una tendencia clara de aumento o disminución del consumo según el tipo de clima; las medianas se mantienen estables.

# **DISTRIBUCION GEOGRAFICA**

```python
def crear_mapa(df, variable, color_densidad, titulo):
  fig = px.choropleth(
    df,
    locations='Country',
    locationmode='country names',
    color=color_densidad,
    hover_name='Country',
    animation_frame='Year',
    color_continuous_scale='Reds',
    title=titulo
  )
  fig.show()

def crear_barras(df, variable_x, variable_y, titulo):
  plt.figure(figsize=(20, 6))
  ax = sns.barplot(data=df, x=variable_x, y=variable_y, palette='dark:red', hue=variable_x)

  for i in ax.containers:
      ax.bar_label(i, fmt='%.0f', label_type='edge', padding=3)
  plt.xticks(rotation=90)
  plt.title(titulo)
  plt.xlabel("Country")
  plt.tight_layout()
  plt.show()
```

# **DISTRIBUCION GEOGRAFICA DEL CONSUMO TOTAL DE AZUCAR ENTRE AMERICA, ASIA Y EUROPA**

```python
mapa_consumo = df[(df['Continent'] == 'America') | (df['Continent'] == 'Asia') | (df['Continent'] == 'Europe')][['Continent', 'Country', 'Year', 'Total_Sugar_Consumption']].dropna()

df_mapa_consumo = mapa_consumo.groupby(['Continent', 'Country', 'Year'], as_index=False)['Total_Sugar_Consumption'].mean()

crear_mapa(df_mapa_consumo, 'Total_Sugar_Consumption', 'Total_Sugar_Consumption', 'Consumo total de azucar a lo largo del tiempo')

ordenar_consumo = df_mapa_consumo.groupby(['Continent', 'Country'], as_index=False)['Total_Sugar_Consumption'].mean().sort_values(by='Total_Sugar_Consumption', ascending=False)

crear_barras(ordenar_consumo, 'Continent', 'Total_Sugar_Consumption', 'Consumo total de azucar en toneladas')
crear_barras(ordenar_consumo, 'Country', 'Total_Sugar_Consumption', 'Consumo total de azucar en toneladas')


```

# **DISTRIBUCION GEOGRAFICA DEL TASA DE OBESIDAD ENTRE AMERICA Y ASIA**

```python
mapa_obesidad = df[(df['Continent'] == 'America') | (df['Continent'] == 'Asia') | (df['Continent'] == 'Europe')][['Country', 'Year', 'Obesity_Rate']].dropna()

df_mapa_obesidad = mapa_obesidad.groupby(['Country', 'Year'], as_index=False)['Obesity_Rate'].mean()

crear_mapa(df_mapa_obesidad, 'Obesity_Rate', 'Obesity_Rate', 'Tasa de obesidad a lo largo del tiempo')

ordenar_obesidad = df_mapa_obesidad.groupby('Country', as_index=False)['Obesity_Rate'].mean().sort_values(by='Obesity_Rate', ascending=False)

crear_barras(ordenar_obesidad, 'Country', 'Obesity_Rate', 'Tasa de obesidad')
```
